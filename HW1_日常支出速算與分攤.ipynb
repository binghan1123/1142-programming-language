```python
import pandas as pd
# read data and put it in a dataframe
# 在 google 工作表載入 gsheets
gsheets = gc.open_by_url('https://docs.google.com/spreadsheets/d/1yefD7L8LSK27T-B_V4PWyk6tEUl5IGUCs9Lw8SbCGF0/edit?usp=sharing')
```

```python
# 從 gsheets 的 All-whiteboard-device 載入 sheets
sheets = gsheets.worksheet('工作表1').get_all_values()
# 將 sheets1 資料載入 pd 的 DataFrame 進行分析
df = pd.DataFrame(sheets[1:], columns=sheets[0])
# 取得最前面的5筆資料
df.head()
```

```python
import datetime
```

```python
# 讓使用者輸入資料
date_str = input("請輸入日期 (格式：YYYY-MM-DD): ")
# 檢查日期格式是否正確
datetime.datetime.strptime(date_str, '%Y-%m-%d')

time_str = input("請輸入時間 (格式：HH:MM): ")
# 檢查時間格式是否正確
datetime.datetime.strptime(time_str, '%H:%M')

kind = input("請輸入分類: ")

item = input("請輸入品項: ")

unit_price = float(input("請輸入單價: "))

amount = float(input("請輸入數量: "))

place = input("請輸入地點: ")

payment_method = input("請輸入支付方式: ")

remark = input("請輸入備註: ")
```

```python
import pandas as pd
# 創建一個新的 DataFrame 來代表你要新增的資料
new_data = pd.DataFrame([{
    '日期': date_str,
    '時間': time_str,
    '分類': kind,
    '品項': item,
    '單價': unit_price, # 修正欄位名稱，移除冒號
    '數量': amount,     # 修正欄位名稱，移除冒號
    '總金額': amount * unit_price, # 修正欄位名稱，移除冒號
    '付款人': "熊本熊本熊",
    '地點': place,
    '支付方式': payment_method,
    '備註': remark
}])

# 使用 concat() 將新資料合併到舊的 df 中
df = pd.concat([df, new_data], ignore_index=True)
```

```python
# 預期的正確欄位名稱（來自原始 Google Sheet）
expected_columns = ['日期', '時間', '分類', '品項', '單價', '數量', '總金額', '付款人', '地點', '支付方式', '備註']

# 處理有重複且名稱不同的欄位
# 例如 '單價' 和 '單價:'，我們希望將數據合併到正確的 '單價' 欄位
columns_to_merge = {
    '單價': ['單價:'],
    '數量': ['數量:'],
    '總金額': ['總金額:'],
    '支付方式': ['付款方式'] # 處理 '支付方式' 和 '付款方式' 的情況
}

for col, old_cols in columns_to_merge.items():
    for old_col in old_cols:
        if old_col in df.columns and col in df.columns:
            # 將 old_col 的非 NaN 值填充到 col 的 NaN 位置
            df[col] = df[col].fillna(df[old_col])
        elif old_col in df.columns and col not in df.columns: # If correct column was somehow dropped, rename
            df = df.rename(columns={old_col: col})

# 移除所有不再需要的額外或重複的欄位
# 首先找出所有不在 expected_columns 裡面的欄位
columns_to_drop = [col for col in df.columns if col not in expected_columns]

# 安全地刪除這些欄位
if columns_to_drop:
    df = df.drop(columns=columns_to_drop)

# 確保 df 的欄位順序與 expected_columns 一致
df = df[expected_columns].copy()

print("DataFrame 已清理完畢，請檢查新的欄位結構：")
display(df.head())
```

```python
# 將 DataFrame 轉換成列表的列表 (list of lists)，這是 gspread 支援的資料格式
# 修改這裡，只寫入 new_data 的內容
data_to_write = new_data.values.tolist()

# 第一步：取得工作表物件
worksheet = gsheets.worksheet('工作表1')

# 第二步：將資料寫入工作表
# append_rows 會返回一個包含更新資訊的字典
response = worksheet.append_rows(values=data_to_write, value_input_option='USER_ENTERED')
print("資料已成功寫入 Google 工作表！")

# 第三步：獲取新寫入資料的範圍並應用置中格式
if 'updates' in response and 'updatedRange' in response['updates']:
    updated_full_range = response['updates']['updatedRange']
    # 從完整的範圍字串中提取 A1 範圍表示法 (例如 'A8:K13')
    # updated_full_range 可能是 "'工作表1'!A8:K13"，我們需要移除工作表名稱部分
    if '!' in updated_full_range:
        updated_range_a1 = updated_full_range.split('!')[1].strip("'")
    else:
        updated_range_a1 = updated_full_range.strip("'")

    worksheet.format(updated_range_a1, {
        'horizontalAlignment': 'CENTER' # 設定水平置中
    })
    print("新寫入的資料已設定為置中對齊。")
else:
    print("無法獲取寫入資料的範圍，因此無法設定置中對齊。")
```
